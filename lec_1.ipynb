```python
# pie_chart
import pandas as pd
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
import plotly.express as px

data={
    "Department":["It","Hr","Finance","Marketing"],
    "Employee":[40,20,25,15]
}

df=pd.DataFrame(data)
fig=px.pie(df,values="Employee",names="Department",color_discrete_sequence=["blue","green","orange","red"],title="Employee Distribution by Department")
fig.show()
```

```python
# pie_chart
import pandas as pd
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
import plotly.express as px

data={
    "Department":["It","Hr","Finance","Marketing"],
    "Employee":[40,20,25,15]
}

df=pd.DataFrame(data)
fig=px.pie(df,values="Employee",names="Department",color_discrete_sequence=["blue","green","orange","red"],title="Employee Distribution by Department")
fig.update_traces(
    textinfo="percent+label",textposition="inside",pull=[0,0.1,0,0]
)
fig.show()
```

```python
# Histogram
import pandas as pd
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
import plotly.express as px

data = {
    "Marks": [40, 45, 50, 55, 60, 65, 70, 75, 80, 85],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
}

df = pd.DataFrame(data)

fig = px.histogram(
    df,
    x="Marks",
    nbins=10,
    color="Gender",
    histfunc="count",
    pattern_shape="Gender",
    text_auto=True
)

fig.show()
```
